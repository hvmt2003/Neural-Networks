```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('/content/placement_cgpa_score_placed.csv')
```

```python
dataset.head()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='cgpa',y='resume_score',data=dataset,hue='placed')
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import Perceptron
pr=Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test)*100
```

```python
pr.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
